**Background of COVID-19**

COVID-19, also known as the coronavirus disease, is caused by a virus called SARS-CoV-2. The disease was first identified in Wuhan, China in December 2019 and quickly spread around the world, leading the World Health Organization (WHO) to declare it a global pandemic in March 2020.

Mexico had been hit hard by the COVID-19 pandemic. It has been one of the countries with the highest number of COVID-19 deaths in the world. Mexico's government had implemented measures such as social distancing, limits on public gatherings, and the closure of non-essential businesses to slow the spread of the virus. However, some experts have criticized the government's response as insufficient and slow.

**Data source**

```python
%matplotlib inline
import pandas as pd 
!pip install xlrd

df = pd.read_excel("s3://reilly-241-2023-python/covid_data.xls") 
df.head()

```

## Q1. How does cases in Mexico compare to America?


```python
Mexico= df.loc[df['countriesAndTerritories']=='Mexico']
US= df.loc[df['countriesAndTerritories']=='United_States_of_America']
```

```python
Mexico.plot.scatter(x='month',y='cases')
```

```python
US.plot.scatter(x='month',y='cases')
```

When comparing both scatter plots, the United States and Mexico have similar COVID-19 cases for the first 9 months of the year. Both countries have a small peak in July (month 7) going to about 125,000 cases however the number of cases goes back downhill in the following months (August and September). While Mexico’s cases stay relatively consistent, its neighbor, the United States has a very high spike in cases starting in October(month 10), doubling the number of cases in Mexico.


## Q2. What month did Mexico have the most cases?


```python
sum_cases_per_month = Mexico.groupby('month').sum()['cases']
sum_cases_per_month.plot()
```

Mexico experienced its highest number of cases in July (month 7) with over 200,000 cases. The country had maintained a relatively low number of cases from January (month 1) until April (month 4) where there was a sharp incline in the number of cases until July. Cases were then decreasing after July until October (month 10) and November (month 11) when there was another peak at just under 200,000 cases.

## Q3. What month did Mexico have the most deaths?


```python
sum_deaths_per_month = Mexico.groupby('month').sum()['deaths']
sum_deaths_per_month.plot()
```

The line graph above represents the amount of COVID-19 deaths in Mexico. The death count is relatively low in the beginning month of the year however significantly increased starting in April (month 4) and peaks in July (month 7). This is consistent with the number of cases in Mexico, in which there was a peak in July. There is a rapid climb of deaths through the months of April (month 4) to July (month 7) and a slower decline starting in August (month 8). The death count is visualized to begin rapidly decreasing as the year closes out.


**Conclusion and Limitations**

The United States and Mexico had similar COVID-19 cases for the first 9 months of the year, however Mexico’s cases stayed relatively consistent while the United States had a high spike in cases towards the end of the calendar year. This spike in cases in the United States is likely due to the laxed travel restrictions in the United States at this time. Mexico experienced its highest number of cases in July (month 7) with over 200,000 cases with additional highs near October and November, as did many countries due to travel during these holiday months. This is also reflected in the highest number of deaths per month in July for Mexico and peaks again in October and November.
